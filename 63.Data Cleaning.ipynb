```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Nahye\OneDrive - University of Toronto\Documents\Alex Analyst\Python_Pandas\Customer Call List.xlsx")
df
```

```python
# remove a duplicate row
df = df.drop_duplicates()
df
```

```python
# delete a column
df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
# Clean last name
df['Last_Name'].str.strip()
# only took out white space
```

```python
df['Last_Name'].str.lstrip("...")
```

```python
df['Last_Name'].str.lstrip("/")
# ... is still there
```

```python
df['Last_Name'].str.lstrip(["/", '...'])
# The argument has to be a single string
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('./_')
df
# Make sure to save the change to df['Last_Name'], not df
# use strip('./_') to strip off those characters from both sides
```

```python
# clean of the phone number
df['Phone_Number'].str.strip('|/')
# didn't work because it removes those characters at the both ends, not in the middle of a str
```

```python
# remove anything other than alphabets(lower or upper)
# [^...] means anything other than ...
# make sure to set regex = True
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex = True)
df
```

```python
# Format numbers to xxx-xxx-xxxx

# df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
# error because some entries are float and the others are str.
# can index str only

# convert all entries to str
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df
```

```python
# Formatting xxx-xxx-xxxx
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
# Remove Na
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df
```

```python
# Clean Address
# Split the address based on ','
df['Address'].str.split(',', expand = True)
```

```python
# Create Street_Address, State and Zip_Code with the split address and append
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', expand = True)
df
```

```python
# drop the column Address (old address form)
df = df.drop(columns = 'Address')
df
```

```python
# Standardize Paying Customer
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
# Remove all N/a from df
df = df.replace('N/a', '')
df
# when looking at df, don't specify str (df.str.replace doesn't work but df.replace does.)
```

```python
# Remove all NaN and None
# NaN and None are not strings, so df = df.replace('NaN', '') doesn't work

df = df.fillna('')
df
```

```python
df.index
```

```python
# Remove Blank Phone_Number and Y for Do_Not_Contact

# Loop through every index/row (defined as x in the code below)
# Go to the Do_Not_Contact and Phone_Number entries of the row using df.loc[x, 'Do_Not_Contact'] and df.loc[x, 'Phone_Number']
# If the entry is Y, remove the row using drop the row

for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y' or df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)

df

# Can remove blank Phone_Number using dropna
# df = df.dropna(subset = 'Phone_Number'), inplace = True)
```

```python
# Re-assign index. set drop = True to remove the old index
df = df.reset_index(drop = True)
df
```
